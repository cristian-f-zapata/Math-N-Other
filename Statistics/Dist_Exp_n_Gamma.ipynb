En una investigacin biomdica se determin 
que el tiempo de supervivencia, en semanas, de un animal cuando se le somete a cierta exposi cin de radiacin gamma tiene una distribucin gamma con  = 5 y 
 = 10. 
 a) Cul es el tiempo medio de supervivencia de un 
animal seleccionado al azar del tipo que se utiliz 
en el experimento?
 b) Cul es la desviacin estndar del tiempo de supervivencia?
 c) Cul es la probabilidad de que un animal sobreviva ms de 30 semanas?

```python
import scipy.stats as stats

# Parmetros de la distribucin gamma
alpha = 5  # Parmetro de forma
beta = 10  # Parmetro de escala

# a) Tiempo medio de supervivencia
tiempo_medio = alpha * beta

# b) Desviacin estndar del tiempo de supervivencia
desviacion_estandar = (alpha * beta**2)**0.5

# c) Probabilidad de que un animal sobreviva ms de 30 semanas
probabilidad_sobrevivir_mas_de_30_semanas = 1 - stats.gamma.cdf(30, a=alpha, scale=beta)

print("a) Tiempo medio de supervivencia:", tiempo_medio, "semanas")
print("b) Desviacin estndar del tiempo de supervivencia:", desviacion_estandar, "semanas")
print("c) Probabilidad de que un animal sobreviva ms de 30 semanas:", probabilidad_sobrevivir_mas_de_30_semanas)
```

Se sabe que la vida, en semanas, de cierto tipo 
de transistor tiene una distribucin gamma con una media de 10 semanas y una desviacin estndar de 50
semanas. 
 a) Cul es la probabilidad de que un transistor de 
este tipo dure a lo sumo 50 semanas?
 b) Cul es la probabilidad de que un transistor de 
este tipo no sobreviva las primeras 10 semanas?

```python
"""esta no diooooooooooooooooooooooooooo"""


import scipy.stats as stats
import numpy as np

# Media y desviacin estndar dados
media = 10  # Media en semanas
desviacion_estandar = np.sqrt(50)  # Desviacin estndar en semanas

# Encontrar los parmetros alfa y beta de la distribucin gamma
alfa = (media ** 2) / desviacion_estandar
beta = desviacion_estandar / media

# a) Probabilidad de que un transistor dure a lo sumo 50 semanas
probabilidad_dure_50_o_menos = stats.gamma.cdf(50, a=alfa, scale=beta)

# b) Probabilidad de que un transistor no sobreviva las primeras 10 semanas
probabilidad_no_sobreviva_10 = stats.gamma.cdf(10, a=alfa, scale=beta)

print("a) Probabilidad de que un transistor dure a lo sumo 50 semanas:", probabilidad_dure_50_o_menos)
print("b) Probabilidad de que un transistor no sobreviva las primeras 10 semanas:", probabilidad_no_sobreviva_10)
```
