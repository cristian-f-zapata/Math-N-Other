En una investigación biomédica se determinó 
que el tiempo de supervivencia, en semanas, de un animal cuando se le somete a cierta exposi ción de radiación gamma tiene una distribución gamma con α = 5 y 
β = 10. 
 a) ¿Cuál es el tiempo medio de supervivencia de un 
animal seleccionado al azar del tipo que se utilizó 
en el experimento?
 b) ¿Cuál es la desviación estándar del tiempo de supervivencia?
 c) ¿Cuál es la probabilidad de que un animal sobreviva más de 30 semanas?

```python
import scipy.stats as stats

# Parámetros de la distribución gamma
alpha = 5  # Parámetro de forma
beta = 10  # Parámetro de escala

# a) Tiempo medio de supervivencia
tiempo_medio = alpha * beta

# b) Desviación estándar del tiempo de supervivencia
desviacion_estandar = (alpha * beta**2)**0.5

# c) Probabilidad de que un animal sobreviva más de 30 semanas
probabilidad_sobrevivir_mas_de_30_semanas = 1 - stats.gamma.cdf(30, a=alpha, scale=beta)

print("a) Tiempo medio de supervivencia:", tiempo_medio, "semanas")
print("b) Desviación estándar del tiempo de supervivencia:", desviacion_estandar, "semanas")
print("c) Probabilidad de que un animal sobreviva más de 30 semanas:", probabilidad_sobrevivir_mas_de_30_semanas)
```

Se sabe que la vida, en semanas, de cierto tipo 
de transistor tiene una distribución gamma con una media de 10 semanas y una desviación estándar de √50
semanas. 
 a) ¿Cuál es la probabilidad de que un transistor de 
este tipo dure a lo sumo 50 semanas?
 b) ¿Cuál es la probabilidad de que un transistor de 
este tipo no sobreviva las primeras 10 semanas?

```python
"""esta no diooooooooooooooooooooooooooo"""


import scipy.stats as stats
import numpy as np

# Media y desviación estándar dados
media = 10  # Media en semanas
desviacion_estandar = np.sqrt(50)  # Desviación estándar en semanas

# Encontrar los parámetros alfa y beta de la distribución gamma
alfa = (media ** 2) / desviacion_estandar
beta = desviacion_estandar / media

# a) Probabilidad de que un transistor dure a lo sumo 50 semanas
probabilidad_dure_50_o_menos = stats.gamma.cdf(50, a=alfa, scale=beta)

# b) Probabilidad de que un transistor no sobreviva las primeras 10 semanas
probabilidad_no_sobreviva_10 = stats.gamma.cdf(10, a=alfa, scale=beta)

print("a) Probabilidad de que un transistor dure a lo sumo 50 semanas:", probabilidad_dure_50_o_menos)
print("b) Probabilidad de que un transistor no sobreviva las primeras 10 semanas:", probabilidad_no_sobreviva_10)
```
