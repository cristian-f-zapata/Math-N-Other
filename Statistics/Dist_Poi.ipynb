Se sabe que la probabilidad de que un estudiante  de preparatoria no pase la prueba de escoliosis (curvatura de la espina dorsal) es de 0.004. De los siguientes 1875 estudiantes que se revisan en bsqueda de escoliosis, Calcule la probabilidad de que:

a) menos de 5 no pasen la prueba;

```python
import math

poisson= lambda lmbda, k: ( math.exp(-lmbda) * lmbda ** k) / math.factorial(k)


lmbda = 0.004*1875 
k = 5


probabilidad= sum(poisson(lmbda, i) for i in range(k))


print(probabilidad)
```
