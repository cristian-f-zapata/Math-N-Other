## a

```python
t_alpha, grados_l=2.069,23
Alpha = 1 - stats.t.cdf(t_alpha, grados_l)
print(f"El nivel de confianza es: {Alpha}")

alpha, grados_l=1-Alpha-0.965, 23
valor_t = stats.t.ppf(1 - alpha, grados_l)
print("Valor de k es:", valor_t, "=t_", 1-Alpha-0.965 )
```

## b

```python
t_alpha, grados_l=2.807,23
Alpha = 1 - stats.t.cdf(t_alpha, grados_l)
print(f"El nivel de confianza es: {Alpha}")

alpha, grados_l = Alpha + 0.095, 23
valor_t = stats.t.ppf(1 - alpha, grados_l)
print("Valor de k es:", valor_t, "=t_", Alpha + 0.095 )
```

## c

Si dentro del intrvalo P(-k<T<k)=0.9, quiere decir que la suma de las reas no contempladas corresponde a 1-0.9=0.1. A este punto es evidente que si se tiene dos de esas reas, cada una tendra una cantidad de 0.05. De modo que k=$t_{0.05}$

```python
alpha, grados_l = 0.05,23
valor_t = stats.t.ppf(1 - alpha, grados_l)
print("Valor de k es:", valor_t, "=t_", Alpha )
```
