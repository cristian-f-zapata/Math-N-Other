A) Demostrar que $\Gamma (1/2) =\sqrt{pi}$

B) Demostrar que  $\Gamma (p+1) = p\Gamma (p) $

C) Demostrar que $\Gamma (p) = (p-1)!$

## A)

$ \Gamma (1/2) =  \int_{0}^{\infty } x^{-1/2} e^{-x} dx $

Usando la sustitucion: 

$u=x^{1/2}, du= 1/2 x^{-1/2} dx, dx=2x^{1/2}du  $


$\Gamma (1/2) = \int_{0}^{\infty }  2e^{-u^{2}}  du$

$\Gamma (1/2) =\sqrt{pi} \int_{0}^{\infty } \frac{2e^{-u^{2}}}{\sqrt{pi}}   du=\sqrt{pi} Erf(x)$

$\Gamma (1/2) =\sqrt{pi}$

## B)

$\Gamma (p+1) =  \int_{0}^{\infty } x^{p} e^{-x} dx$

Usando la sustitucion: 

$u=x^{p}, du= px^{p-1}dx  $


$dv=e^{-x}, v= -e^{-x}dx  $

$\Gamma (p+1) =  - x^{p} e^{-x} + \int_{0}^{\infty } px^{p-1} e^{-x}$

$\Gamma (p+1) =p \int_{0}^{\infty }  x^{p-1} e^{-x}= p\Gamma (p) $

## C)

$\Gamma (p) = (p-1) \Gamma (p-1) = (p-1)(p-2)\Gamma (p-2) = (p-1)(p-2)(p-3)\Gamma (p-3)=p(p-1)(p-2)... 1$

$\Gamma (p) = (p-1)!$

## Distribucion Beta

mayores a 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import beta, gamma

# Valores de lambda y r
lambda_values = [2, 4, 3, 5]
r_values = [5, 3, 4, 2]

# Rango de valores de x
x = np.linspace(0, 1, 100)

# Graficar la función beta para cada par de valores de lambda y r
for i in range(len(lambda_values)):
    lambda_val = lambda_values[i]
    r_val = r_values[i]
    
    y = beta(lambda_val, r_val) * x**(lambda_val - 1) * (1 - x)**(r_val - 1)
    
    plt.plot(x, y, label=f'λ={lambda_val}, r={r_val}')

# Configuración de la gráfica
plt.title('Función Beta para diferentes valores de λ y r')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```

## Weibull

El diámetro de ejes de acero sigue la distribución de weibull con parámetros gama=1, beta=2 y delta=0.5. Encuentre la probabilidad de que en un eje seleccionado al azar no excederá 1.5 pulgadas de diámetro

```python
from scipy.stats import weibull_min

# Parámetros de la distribución Weibull
gamma = 1
beta = 2
delta = 0.5

# Diámetro máximo deseado
x = 1.5

# Cálculo de la probabilidad P(X <= 1.5)
probability =( 1-np.exp(- ( (x-gamma ) / delta )**beta ))

print(f"La probabilidad de que un eje no exceda 1.5 pulgadas de diámetro es: {probability:.4f}")
```

Se sabe que el tiempo de falla de cierto transistor sigue la distribución de weibull con parámetros gama =0 beta = 1/3 y delta=400. obtenga la facción esperada que durará 600 horas

```python
# Parámetros de la distribución Weibull
gamma = 0
beta = 1/3
delta = 400

# Tiempo mínimo deseado
t_min = 600

# Cálculo de la fracción esperada P(T >= 600)
probability =1-( 1 - np.exp(-((t_min - gamma) / delta) ** beta))

print(f"La fracción esperada de tiempo que dura 600 horas o más es: {probability:.4f}")
```

grafique la funcion de weibull con gama = 0 delta=1 y beta=1,3,
 y 4

```python
# Parámetros fijos
gamma = 1  # También conocido como beta
delta = 1

# Valores de x
x = np.linspace(1, 5, 1000)

# Función de Weibull para diferentes valores de beta (gamma)
betas = [1, 2, 3, 4]

plt.figure(figsize=(10, 6))
for beta in betas:
    pdf = (beta / delta) * ( ( (x -gamma) / delta ) ** (beta - 1) ) * np.exp(-( ((x - gamma)/ delta) ** beta))
    plt.plot(x, pdf, label=f'Beta={beta}')

plt.title('Función de Weibull para diferentes valores de Beta ')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

En 3D

```python
from mpl_toolkits.mplot3d import Axes3D

# Parámetros fijos
gamma = 1
delta = 1

# Valores de x y beta
x = np.linspace(1, 5, 100)
betas = np.linspace(1, 5, 100)

# Crear una cuadrícula de valores de x y beta
X, Beta = np.meshgrid(x, betas)

# Calcular la función de Weibull para diferentes valores de beta (gamma)
pdf = (Beta / delta) * (((X - gamma) / delta) ** (Beta - 1)) * np.exp(-(((X - gamma) / delta) ** Beta))

# Crear una figura 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie 3D
ax.plot_surface(X, Beta, pdf, cmap='viridis')

# Etiquetas de los ejes
ax.set_xlabel('x')
ax.set_ylabel('Beta ')
ax.set_zlabel('Densidad de probabilidad')

# Título
plt.title('Función de Weibull en 3D para diferentes valores de Beta ')

# Mostrar la figura
plt.show()
```

la densidad de tiempo de falla corresponde a un sistema de computadora pequeño tiene un densidad de weibull con beta=1/4, gama=0 y delta=200, ¿que fraccion de estas unidades durara mas de 1000 horas?

```python
# Parámetros de la distribución de Weibull
beta = 1/4
gamma = 0
delta = 200

# Valor de interés (duración deseada)
x = 1000

# Calcular la probabilidad usando la función de supervivencia de Weibull
probabilidad = 1-( 1-np.exp(- ( (x-gamma ) / delta )**beta ))

# Imprimir la probabilidad
print(f"La probabilidad de que una unidad dure más de {x} horas es aproximadamente {probabilidad:.4f}")
```

```python
from scipy.special import gamma

# Parámetros de la distribución de Weibull
beta = 1/4
delta = 200

# Calcular el MTTF
mttf = delta * gamma(1 + 1/beta)

# Imprimir el MTTF
print(f"El tiempo medio de falla (MTTF) es aproximadamente {mttf:.2f} horas")
```

Suponga que la variable aleatoria X tiene una 
distribución beta con α = 1 y β = 3.
 a) Determine la media y la mediana de X.
 b) Determine la varianza de X.
 c) Calcule la probabilidad de que X > 1/3

```python
import scipy.stats as stats

# Parámetros de la distribución beta
alpha = 1
beta = 3

# a) Media y mediana de X
media = stats.beta.mean(alpha, beta)
mediana = stats.beta.median(alpha, beta)

print("a) Media de X:", media)
print("   Mediana de X:", mediana)
```

```python
varianza = stats.beta.var(alpha, beta)
print("b) Varianza de X:", varianza)
```

```python
probabilidad = 1 - stats.beta.cdf(1/3, alpha, beta)
print("c) Probabilidad de que X > 1/3:", probabilidad)
```

Si la proporción de una marca de televisores que 
requiere servicio durante el primer año de operación es 
una variable aleatoria que tiene una distribución beta 
con α = 3 y β = 2, ¿cuál es la probabilidad de que al 
menos 80% de los nuevos modelos de esta marca que 
se vendieron este año requieran servicio durante su primer año de operación?

```python
import scipy.stats as stats

# Parámetros de la distribución beta
alpha = 3
beta = 2

# Probabilidad de que X < 0.80
probabilidad_menor_a_0_80 = stats.beta.cdf(0.80, alpha, beta)

# Probabilidad de que X >= 0.80
probabilidad_al_menos_0_80 = 1 - probabilidad_menor_a_0_80

print("La probabilidad de que al menos el 80% requiera servicio es:", probabilidad_al_menos_0_80)
```
