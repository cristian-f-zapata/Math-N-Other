```python
from scipy.stats import ttest_1samp
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
from scipy.stats import norm
from scipy.stats import t
import numpy as np
import math
```

```python
# Datos
media_muestral = 5.23
media_hipotesis_nula = 5.5
tamanio_muestra = 64
grados_libertad = tamanio_muestra - 1
nivel_significancia = 0.05
desviacion_estandar_poblacional = 0.24

# Calcular el estadístico de prueba z
z_statistic = (media_muestral - media_hipotesis_nula) / (desviacion_estandar_poblacional / math.sqrt(tamanio_muestra))

# Calcular el valor crítico para prueba z de cola izquierda
z_critico = norm.ppf(nivel_significancia)

# Imprimir el estadístico de prueba z
print("Estadístico de prueba z:", z_statistic)

# Imprimir el valor crítico y tomar una decisión
print("Valor crítico para prueba z de cola izquierda:", z_critico)
if z_statistic < z_critico:
    print("Se rechaza la hipótesis nula. Hay evidencia suficiente para afirmar que la media es menor a 5.5 onzas.")
else:
    print("No se rechaza la hipótesis nula. No hay evidencia suficiente para afirmar que la media es menor a 5.5 onzas.")
```

como se tiene la desviacion_estandar_poblacional y no de l muestra se usa Z


```python
media_pob = 800
desviacion_estandar_poblacional = 40
tamanio_muestra = 30
promedio_muestra = 788
nivel_significancia=0.05

# Calcular el estadístico de prueba z
z_statistic = (promedio_muestra - media_pob) / (desviacion_estandar_poblacional / (tamanio_muestra**0.5))

# Calcular el valor p (probabilidad de obtener un resultado al menos tan extremo como el observado)
p_valor = 2 * (1 - norm.cdf(abs(z_statistic)))

# Imprimir el valor P
print("Valor P:", p_valor,"z", z_statistic)

if p_valor < nivel_significancia:
    print(f"Rechazamos la hipótesis nula. Hay evidencia suficiente para afirmar que la media es diferente que {media_pob}.")
else:
    print(f"No podemos rechazar la hipótesis nula. No hay suficiente evidencia para afirmar que  la media es diferente de {media_pob}.")
```
