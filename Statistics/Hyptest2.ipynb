```python
from scipy.stats import f
from scipy.stats import t
import numpy as np
from scipy.stats import ttest_rel
```

H0 : ^2 A = ^2 B, H1 : ^2 A != ^2 B


```python
instrumento_A = np.array([0.86, 0.82, 0.75, 0.61, 0.89, 0.64, 0.81, 0.68, 0.65])
instrumento_B = np.array([0.87, 0.74, 0.63, 0.55, 0.76, 0.70, 0.69, 0.57, 0.53])

# Calcula las varianzas de las dos muestras
varianza_A = np.var(instrumento_A, ddof=1)
varianza_B = np.var(instrumento_B, ddof=1)

# Calcula el estadstico F y el valor p para la prueba F
estadistico_F = varianza_B / varianza_A
grados_libertad_A = len(instrumento_A) - 1
grados_libertad_B = len(instrumento_B) - 1
valor_p = 2 * min(f.cdf(estadistico_F, grados_libertad_A, grados_libertad_B), 1 - f.cdf(estadistico_F, grados_libertad_A, grados_libertad_B))

print("Valor p y estadistico F:", valor_p, "," ,estadistico_F)

nivel_significancia = 0.05
if valor_p < nivel_significancia:
    print("Rechazamos la hiptesis nula. Hay evidencia suficiente para afirmar que las varianzas de las mediciones de los dos instrumentos no son iguales.")
else:
    print("No podemos rechazar la hiptesis nula. No hay suficiente evidencia para afirmar que las varianzas de las mediciones de los dos instrumentos son diferentes.")
```

```python
precirugia = np.array([10.80, 12.90, 9.59, 8.81, 12.00, 6.07])  # Leucocitos GB preciruga
postcirugia = np.array([10.60, 16.60, 17.20, 14.00, 10.60, 8.60])  # Leucocitos GB postciruga
n = len(precirugia) # Calcular el tamao de la muestra
nivel_significancia = 0.05

# Calcular las diferencias
diferencias = precirugia - postcirugia

# Calcular la media de las diferencias
media_diferencias = np.mean(diferencias)

# Calcular la desviacin estndar de las diferencias
desviacion_diferencias = np.std(diferencias, ddof=1)

# Calcular el estadstico de prueba t
t_statistic = media_diferencias / (desviacion_diferencias / np.sqrt(n))

# Calcular los grados de libertad
grados_libertad = n - 1

# Calcular el valor crtico para la prueba t de dos colas
t_critico = t.ppf(1 - nivel_significancia/2, df=grados_libertad)

# Imprimir el estadstico de prueba t
print("Estadstico de prueba t:", t_statistic)

# Imprimir el valor crtico y tomar una decisin
print("Valor crtico para prueba t de dos colas:", t_critico)
if abs(t_statistic) > t_critico:
    print("Hay evidencia suficiente para rechazar la hiptesis nula.")
else:
    print("No hay suficiente evidencia para rechazar la hiptesis nula.")
```
