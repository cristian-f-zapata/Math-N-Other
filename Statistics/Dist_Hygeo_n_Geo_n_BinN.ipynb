El dueo de una casa planta 6 bulbos seleccionados al azar de una caja que contiene 5 bulbos de tulipn 
y 4 de narciso. Cul es la probabilidad de que plante 2 
bulbos de narciso y 4 de tulipn?




```python
import math

# Parmetros del problema
N1 = 4
N2 = 5 # N2=N-K
N = 9
n = 6
x = 2

probability = (math.comb(N1, x) * math.comb(N2, n - x)) / math.comb(N, n)

print(f"La probabilidad de plantar 2 bulbos de narciso y 4 de tulipn es: {probability:.4f}")
```

Se selecciona al azar un comit de 3 personas a partir de 4 mdicos y 2 enfermeras. Escriba una 
frmula para la distribucin de probabilidad de la variable aleatoria X que representa el nmero de mdicos 
en el comit. Calcule P(2  X  3)

```python

N1 = 4 #total medicos
N2 = 2 #total_enfermeras
N=N1+N2
n = 3
 

# Calcular P(X )
hipG=lambda x : (math.comb(N1, x) * math.comb(N2, n-x)) / math.comb(N, n)

# Calcular P(2  X  3) = p(x=2) + p(x=3)

p= sum(hipG(i) for i in range(2, 3 + 1) )

print(f"P(2  X  3) = {p:.4f}")
```
