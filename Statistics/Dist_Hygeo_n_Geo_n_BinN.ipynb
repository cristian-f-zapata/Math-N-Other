El dueño de una casa planta 6 bulbos seleccionados al azar de una caja que contiene 5 bulbos de tulipán 
y 4 de narciso. ¿Cuál es la probabilidad de que plante 2 
bulbos de narciso y 4 de tulipán?




```python
import math

# Parámetros del problema
N1 = 4
N2 = 5 # N2=N-K
N = 9
n = 6
x = 2

probability = (math.comb(N1, x) * math.comb(N2, n - x)) / math.comb(N, n)

print(f"La probabilidad de plantar 2 bulbos de narciso y 4 de tulipán es: {probability:.4f}")
```

Se selecciona al azar un comité de 3 personas a partir de 4 médicos y 2 enfermeras. Escriba una 
fórmula para la distribución de probabilidad de la variable aleatoria X que representa el número de médicos 
en el comité. Calcule P(2 ≤ X ≤ 3)

```python

N1 = 4 #total medicos
N2 = 2 #total_enfermeras
N=N1+N2
n = 3
 

# Calcular P(X )
hipG=lambda x : (math.comb(N1, x) * math.comb(N2, n-x)) / math.comb(N, n)

# Calcular P(2 ≤ X ≤ 3) = p(x=2) + p(x=3)

p= sum(hipG(i) for i in range(2, 3 + 1) )

print(f"P(2 ≤ X ≤ 3) = {p:.4f}")
```

Una empresa está interesada en evaluar su procedimiento de inspección actual para embarques de 50 
artículos idénticos. El procedimiento consiste en tomar 
una muestra de 5 artículos y aceptar el embarque si no 
se encuentran más de 2 defectuosos. ¿Qué proporción 
de embarques con 20% de artículos defectuosos se 
aceptará?

```python

# Parámetros del problema
N = 50
N1 = 10   #50*0.20
N2=N-N1
n = 5


# Calcular la probabilidad de aceptar el embarque (menos de 3 artículos defectuosos en la muestra)
# Calcular P(X )
hipG=lambda x : (math.comb(N1, x) * math.comb(N2, n-x)) / math.comb(N, n)

# Calcular P(2 ≤ X ≤ 3) = p(x=2) + p(x=3)

p= sum(hipG(i) for i in range(0, 2 + 1))

print(f"P(0 ≤ X ≤ 2) = {p:.4f}")
```
