## Dynamics of a Particle in an Infinite Well

Quantum mechanics describes the behavior of particles at the smallest scales, often revealing phenomena that defy classical intuition. This code employs Python's scientific computing libraries, NumPy and Matplotlib, to create a visual representation of the temporal evolution of wave functions for a particle confined within an infinite well.

In quantum mechanics, an infinite well represents a simple yet insightful model where a particle is restricted within a finite region. The length of the well is denoted by `L`. The code illustrates the evolution of wave functions, which are mathematical representations of the probability amplitudes for finding the particle at different positions within the well.

The initial setup involves discretizing the spatial domain (`x`) into 1000 points within the well. The wave functions for different quantum states (`n`) are defined using the sine function. These functions represent distinct energy states of the particle.

A visual representation is created using Matplotlib, showcasing the temporal evolution of the wave functions. The plot displays three states: the fundamental state (n=1) and two excited states (n=2 and n=3). Each state is represented by a distinct line on the graph.

To animate the quantum dynamics, a time-dependent phase factor is introduced to the wave functions. The `update` function iterates through frames, modifying the y-data of each line to depict the changing wave functions over time. The animation consists of 200 frames with a frame interval of 50 milliseconds.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Infinite well parameters
L = 1.0  # Length of the infinite well

# Initial values
x = np.linspace(0, L, 1000)

# Wave functions for different states
def wave_function(x, n, L):
    return np.sqrt(2 / L) * np.sin(n * np.pi * x / L)

# Initialize figure and axis
fig, ax = plt.subplots()
lines = []

# Additional configurations
ax.set_title('Temporal Evolution - Particle in an Infinite Well')
ax.set_xlabel('Position (x)')
ax.set_ylabel('Wave Function')
ax.axhline(0, color='black', linewidth=0.5, linestyle='--', label='x-axis')
ax.legend()
ax.grid(True)

# Set y-axis limits
ax.set_ylim([-2, 2])

# Create lines for each state (fundamental and two excited)
for n in range(1, 4):
    line, = ax.plot(x, wave_function(x, n, L), label=f'State {n}')
    lines.append(line)

# Animation function
def update(frame):
    t = frame * 0.02  # Time parameter
    for n, line in enumerate(lines, start=1):
        line.set_ydata(wave_function(x, n, L) * np.cos(n * np.pi * t))

# Create the animation
num_frames = 200
animation = FuncAnimation(fig, update, frames=num_frames, interval=50, blit=False)

# Save the animation as a GIF file
animation.save('multi_wave_animation.gif', writer='pillow', fps=30)

# Show the static plot
plt.show()

```
